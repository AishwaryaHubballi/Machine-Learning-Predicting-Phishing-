```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv('C:\\Users\\hp\\Downloads\\Phishing Data - Phishing Data.csv') \\READING DATA INTO DATAFRAME
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum() #TO FIND NUMBER OF NULL VALUES IN EACH ATTRIBUTE
```

```python
features = df.iloc[:,1:30]  \\STORING ALL FEATURES IN 'features' 
result = df['Result']   \\\\STORING RESULT ie TARGET VARIABLE IN 'result' 
```

```python
features.head()
```

```python
result.head()
```

```python
#SPLIT THE DATA INTO TRAINING AND TESTING DATA WITH 1% OF DATA AS TEST DATA

from sklearn.model_selection import train_test_split
features_train,features_test,result_train,result_test=train_test_split(features,result,test_size=0.01,random_state=42)
```

```python
#USING GAUSSIAN NAIVE BAYES CLASSIFIER

clf_gnb = GaussianNB()  #CREATE A CLASSIFIER
clf_gnb.fit(features_train,result_train)  #FIT THE TRAINING DATA AND TRAIN THE MODEL
result_pred=clf_gnb.predict(features_test)  #MAKE PREDICTIONS ON TEST DATA
print(classification_report(result_test,result_pred))  #GENERATE THE REPORT ON PREDICTED VALUES
print("accuracy on training set is " + str(clf_gnb.score(features_train,result_train)))  #PRINT ACCURACY
print("accuracy on testing set is " + str(clf_gnb.score(features_test,result_test)))  #PRINT ACCURACY
```

THERE IS DIFFERENCE BETWEEN ACCURACY OF TRAINING SET AND TESTING SET.
HENCE BY CONSIDERING MORE % OF DATA FOR TRAINING, WE ARE TRYING TO OVERFIT OUR MODEL.
HENCE LET US REDUCE THE % OF DATA FOR TRAINING.

```python
#SPLIT THE DATA INTO TRAINING AND TESTING DATA WITH 33% OF DATA AS TEST DATA

from sklearn.model_selection import train_test_split
features_train,features_test,result_train,result_test=train_test_split(features,result,test_size=0.33,random_state=42)
```

```python
#USING GAUSSIAN NAIVE BAYES CLASSIFIER

clf_gnb = GaussianNB()  #CREATE A CLASSIFIER
clf_gnb.fit(features_train,result_train)  #FIT THE TRAINING DATA AND TRAIN THE MODEL
result_pred=clf_gnb.predict(features_test)  #MAKE PREDICTIONS ON TEST DATA
print(classification_report(result_test,result_pred))  #GENERATE THE REPORT ON PREDICTED VALUES
print("accuracy on training set is " + str(clf_gnb.score(features_train,result_train)))  #PRINT ACCURACY
print("accuracy on testing set is " + str(clf_gnb.score(features_test,result_test)))  #PRINT ACCURACY
```

THERE IS NO MUCH DIFFERENCE IN ACCURACIES OF TRAINING AND TESTING SET.
LET US CONSIDER THIS SPLIT FOR TRAINING FURTHER MODELS.

```python
#USING SUPPORT VECTOR CLASSIFIER

clf_svc=SVC(random_state=42)
clf_svc.fit(features_train,result_train)
result_pred=clf_svc.predict(features_test)
print(classification_report(result_test,result_pred))
print("accuracy on training set is " + str(clf_svc.score(features_train,result_train)))
print("accuracy on testing set " + str(clf_svc.score(features_test,result_test)))
```

```python
#USING DECISI0N TREE CLASSIFIER

from sklearn import tree
clf_dt = tree.DecisionTreeClassifier()
clf_dt = clf_dt.fit(features_train,result_train)
result_pred = clf_dt.predict(features_test)
print(classification_report(result_test,result_pred))
print("accuracy is " + str(clf_dt.score(features_test,result_test)))
```

```python
#USING RANDOM FOREST CLASSIFIER

clf_rfc=RandomForestClassifier(n_estimators=10,criterion="entropy",max_features=5,random_state=42)
clf_rfc.fit(features_train,result_train)
result_pred=clf_rfc.predict(features_test)
print(classification_report(result_test,result_pred))
print("accuracy is " + str(clf_rfc.score(features_test,result_test)))
```

```python
#APPLY PCA(PRINCIPLE COMPONENT ANALYSIS) ON TRAINING DATA

pca=PCA(n_components=15)  
pca.fit(features_train)
features_train_pca=pca.transform(features_train)
features_test_pca=pca.transform(features_test)
```

```python
#USING GAUSSIAN NAIVE BAYES CLASSIFIER AFTER APPLYING PCA

clf_gnb_pca = GaussianNB()
clf_gnb_pca.fit(features_train_pca,result_train)
result_pred = clf_gnb_pca.predict(features_test_pca)
print(classification_report(result_test,result_pred))
print("accuracy is " + str(clf_gnb_pca.score(features_test_pca,result_test)))
```

```python
#USING RANDOM FOREST CLASSIFIER AFTER APPLYING PCA

rfc_pca=RandomForestClassifier(criterion="entropy",random_state=42)
rfc_pca.fit(features_train_pca,result_train)
result_pred=rfc_pca.predict(features_test_pca)
print(classification_report(result_test,result_pred))
print("accuracy is " + str(rfc_pca.score(features_test_pca,result_test)))
```

```python
#USING SUPPORT VECTOR CLASSIFIER AFTER APPLYING PCA

svc_pca=SVC(random_state=42)
svc_pca.fit(features_train_pca,result_train)
result_pred=svc_pca.predict(features_test_pca)
print(classification_report(result_test,result_pred))
print("accuracy is " + str(svc_pca.score(features_test_pca,result_test)))
```

```python
#USING DECISION TREE CLASSIFIER AFTER APPLYING PCA

dt_pca=tree.DecisionTreeClassifier()
dt_pca.fit(features_train_pca,result_train)
result_pred=dt_pca.predict(features_test_pca)
print(classification_report(result_test,result_pred))
print("accuracy is " + str(dt_pca.score(features_test_pca,result_test)))
```
